```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
from sklearn import datasets
```

The cost function of linear regression is $$\hat f=\textrm{argmin}_{f} \sum_{i=1}^{N}(y_i-\sum_{j=0}^{p}\beta_j x_{ij})^2$$

The analytical solution to linear regression is: $$\beta = (X^{T}X)^{-1}X^{T}y$$

In the following example, I have generated data points along the line $$y=2x+3, x \in [-1,1]$$

```python
x = np.linspace(-1,1,50)
y = np.array(2 * x + 3 + np.random.normal(size=x.shape[0])).reshape(-1,1)
plt.scatter(x, y)
```

```python
X = np.ones((x.shape[0],2))
X[:,1] = x
X_t = np.transpose(X)
A = np.linalg.inv(np.matmul(X_t, X))
β_analytical = np.matmul(np.matmul(A, X_t), y)
print(f"The analytical solution returned β0={β_analytical[0]}, β1={β_analytical[1]}")
```

```python
def coordinate_descent_linear_regression(β, X, y, num_iter, α):
    '''
        Coordinate descent implementation for linear regression
        α is the step size
    '''
    num_features = X.shape[1]
    β_hist = np.zeros((num_iter, num_features), dtype=float)
    for i in range(num_iter):
        for j in range(num_features):
            gradient = np.dot(X[:,j], (np.matmul(X,β) - y))
            β[j,0] = β[j,0] - α * gradient 
            β_hist[i,j] = β[j,0]
    return [β, β_hist]
```

```python
# Initialize β to be [5,5], setting 
β_init = np.array([5,5], dtype=float).reshape(-1,1)
α = 0.03
num_iter = 50
[β_coord, β_hist] = coordinate_descent_linear_regression(β_init, X, y, num_iter, α)
print(f"The coordinate descent solution returned β0={β_coord[0]}, β1={β_coord[1]}")
```

The cost function of the lasso problem can be written as: 
$$f = \sum_{i=1}^{N}(y_i-\beta_i)^2 +\lambda\sum_{i=1}^{N}|\beta_i| $$
I used the "diabetes" dataset that was included as part of the sklearn package, and obtained the lasso path using the coordinate descent algorithm that I wrote. The paths showed good agreement with the ones obtained using the sklearn package.

```python
def soft_max(ρ, λ):
    '''
        Softmax function
    '''
    sign = 0
    if ρ > 0:
        sign = 1
    elif ρ < 0:
        sign = -1
    return sign * max(0, abs(ρ)-λ)
    
def coordinate_descent_lasso(β, λ, X, y, num_iter, intercept=False):
    '''
        Coordinate descent implementation for solving the Lasso problem
    '''
    num_features = X.shape[1]
    X = X / (np.linalg.norm(X, axis=0)) # normalize X by feature
    β_hist = np.zeros((num_iter, num_features), dtype=float)
    for i in range(num_iter):
        for j in range(num_features):
            Xj = X[:,j].reshape(-1,1)
            y_predict = X @ β
            ρ = Xj.T @ (y - y_predict + β[j,0]*Xj)
            # do not regularize the intercept (constant term)
            if intercept == True:
                β[j,0] = soft_max(ρ, λ) if j != 0 else ρ
            else:
                β[j,0] = soft_max(ρ, λ)
            β_hist[i,j] = β[j,0]
    return [β.flatten(), β_hist]

def plot_result(β_coord, λ_coord, β_sklearn, λ_sklearn, num_features, diabetes):
    fig, ax = plt.subplots(1,2, figsize=(15,7))
    for i in range(num_features):
        ax[0].plot(λ_coord, β_coord[:,i], label=diabetes.feature_names[i])
    ax[0].set_xscale('log')
    ax[0].set_xlabel('log($\\lambda$)')
    ax[0].set_ylabel('Coefficients')
    ax[0].set_title('Lasso paths - Coordinate descent')
    ax[0].legend()
    ax[0].axis('tight')
    
    β_sklearn = β_sklearn[0]
    for i in range(num_features):
        ax[1].plot(λ_sklearn, β_sklearn[i].T, label = diabetes.feature_names[i])
    ax[1].set_xscale('log')
    ax[1].set_xlabel('Log($\\lambda$)')
    ax[1].set_ylabel('coefficients')
    ax[1].set_title('Lasso paths - Sklearn')
    ax[1].legend()
    ax[1].axis('tight')
```

```python
num_iter = 100
diabetes = datasets.load_diabetes()
X = diabetes.data
X = X / np.linalg.norm(X,axis=0)
num_features = X.shape[1]
y = diabetes.target.reshape(-1,1)
# initialize β to be 1s
β_init = np.ones((num_features,1))
λ_coord = np.logspace(0,4,300)/10
β_coord = np.zeros((len(λ_coord),num_features))
for idx, l in enumerate(λ_coord):
    [β, _] = coordinate_descent_lasso(β_init, l, X, y, num_iter, intercept=False)
    β_coord[idx,:] = β
    
eps = 5e-5
λ_sklearn, β_sklearn, _ = linear_model.lasso_path(X, y, eps=eps, fit_intercept=False)
plot_result(β_coord, λ_coord, β_sklearn, λ_sklearn, num_features, diabetes)
```
